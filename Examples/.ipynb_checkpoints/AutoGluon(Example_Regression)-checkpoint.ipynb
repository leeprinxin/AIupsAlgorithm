```python
import numpy as np

def r2_score(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    ss_total = np.sum((y_true - np.mean(y_true))**2)
    ss_residual = np.sum((y_true - y_pred)**2)
    return 1 - (ss_residual / ss_total)

r2_score(new_model.model.predict(X_test), y_test), r2_score(new_model.predict(X_test), y_test)
```

```python
import time
class AutogluonWrapper:
    def __init__(self, predictor, feature_names):
        self.ag_model = predictor
        self.feature_names = feature_names
    
    def predict(self, X):
        if isinstance(X, pd.Series):
            X = X.values.reshape(1,-1)
        if not isinstance(X, pd.DataFrame):
            X = pd.DataFrame(X, columns=self.feature_names)
        return self.ag_model.predict(X)
    
def print_accuracy(f):
    print("Root mean squared test error = {0}".format(np.sqrt(np.mean((f(X_test) - y_test)**2))))
    time.sleep(0.5) # to let the print get out before any progress bars
```

```python
from typing import Union
from sklearn.metrics import log_loss, roc_auc_score, accuracy_score, recall_score, precision_score

extra_metrics = []
for metric_name, metric_func  in __regression_metrics__.items():
    if metric_name == 'auc_score' and y_proba is not None:  
        scorer = make_scorer(
            name=metric_name,
            score_func=metric_func,  # 自定义的评分函数
            optimum=1,
            greater_is_better=True,
            metric_kwargs={"labels": y_test.unique()},  # 传递需要的分类标签
            needs_proba=True  # 因为 roc_auc_score 需要概率
        )
        score=scorer(y_true=y_test, y_pred=y_proba.to_numpy())
        
    elif metric_name in ['precision_score', 'recall_score', 'accuracy_score']:
        scorer = make_scorer(
            name=metric_name,
            score_func=metric_func,
            optimum=1,
            greater_is_better=True
        )
        score=scorer(y_test, y_pred)
    elif metric_name == 'log_loss' and y_proba is not None:
        scorer = make_scorer(
            name=metric_name,
            score_func=metric_func,
            optimum=0,
            greater_is_better=False,
            metric_kwargs={"labels":y_test.unique()}
        )
        score=scorer(y_true=y_test, y_pred=y_proba.to_numpy())
        
    elif metric_name == 'adjust_r2':
        scorer = make_scorer(
            name=metric_name,
            score_func=metric_func,
            optimum=1,
            greater_is_better=True,
            metric_kwargs={ "p":X_test.shape[1]}
        )
        score=scorer(y_test, y_pred)
    elif metric_name == "r2":
        scorer = make_scorer(
            name=metric_name,
            score_func=metric_func,
            optimum=1,
            greater_is_better=True
        )
        score=scorer(y_test, y_pred)
    else:
        scorer = make_scorer(
            name=metric_name,
            score_func=metric_func,
            optimum=0,
            greater_is_better=False
        )
        score=scorer(y_test, y_pred)
    extra_metrics.append(scorer)
    print(metric_name, score)
```
